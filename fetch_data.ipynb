```python
from pybaseball import batting_stats
import pandas as pd
```

```python
selected_cols = ['1','2','3','4','5','6','7', '11', '16', '21', '39', '40', '50', '58', '60', '61', 'DEFENSE', 'FIELDING']
train_data = batting_stats(2000, 2009, qual = 100, stat_columns = selected_cols)
```

```python
train_data.shape
```

```python
selected_cols = ['1','2','3','4','5','6','7', '11', '16', '21', '39', '40', '50', '58', '60', '61', 'DEFENSE', 'FIELDING']
train_data2 = batting_stats(2010, 2019, qual = 100, stat_columns = selected_cols)
```

```python
train_data3 = batting_stats(1980, 1989, qual = 100, stat_columns = selected_cols)
```
